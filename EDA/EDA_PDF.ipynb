```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris = pd.read_csv('iris.csv')
iris
```

```python
iris.columns
```

```python
iris['species'].value_counts()
```

```python
iris.plot(kind = "scatter" , x = 'sepal_length', y = 'sepal_width')
```

```python
sns.FacetGrid(iris, hue = 'species', size = 5)\
    .map(sns.distplot, 'petal_length')\
    .add_legend()
plt.show()
```

-Here the softcurve is Probability Density Function (PDF) . Y-axis is counts . X-axis is Petal Length . 
-Example : It means there are 2.5 setosa flowers of petal_length 1.5.
-Here the curve of versicolor and virginica overlapping.But by we seeing the curve of setosa we can tell all flower which are having petal_length less than 2 are setosa . 
-Petal_length greater than 2 might be versicolor or virginica . 

```python
sns.FacetGrid(iris, hue = 'species', size = 5)\
    .map(sns.distplot, "petal_width")\
    .add_legend()
plt.show()
```

-Here the PDF curve is connected for setosa and versicolor . And also the versicolor and virginica are overlapping so we can say that petal_length is better than petal_width because  it seperated versicolor and virginica .
-So petal_length > petal_width in terms of comparision .

```python
sns.FacetGrid(iris, hue = "species", size = 5)\
    .map(sns.distplot, "sepal_length")\
    .add_legend()
plt.show()
```

-Here we can see all the species are overlapping with each other . so we won't be able to recognize the species . 
-So petal_length > petal_width >>> sepal_length .
